```python
from seaborn.matrix import despine
#sns.set_style("white")
tips = sns.load_dataset('tips')

#errorbar in details
#https://github.com/mwaskom/seaborn/issues/1427
# gap = 0.15 for spacing between the bar
bar = sns.barplot(x=tips.day, y=tips.total_bill, capsize = 0.1, edgecolor='0.1',lw = 1.2,  hue_order=['Female','Male'], errorbar=('se'), hue="sex", data=tips,
                  palette=['red','black'], err_kws={'color': '0.1', 'linewidth': 1.2}, zorder = 5, gap = 0.15);
despine() # Despine should be after plotting the graph


# No use here
#kwargs = {'edgecolor':'0.2','lw':2.5, 'fc':None }

for axis in ['bottom','left']:
  bar.spines[axis].set_linewidth(1.2) #line width of axis line
  bar.spines[axis].set_color('0.2')

#legend loc is the position of legend
plt.legend(frameon = False, prop = {'weight':'bold', 'size':10 }, loc =(0.02, 0.85), ncol=1)

bar.set_xlabel("Day", fontsize = 12, weight='bold', color='0.2')
bar.set_ylabel("Total Bill", fontsize = 12, weight='bold', color='0.2')

#For ticks in the x-axis and y-axis
plt.xticks(size=12, rotation = 45, rotation_mode='anchor', fontsize = 11, ha = 'right', weight = 'bold', color='0.2')
plt.yticks(size=12, weight = 'bold', fontsize = 11, color='0.2')

#Adding asterisk
plt.text(1.15, 23.5, '**', fontsize = 12)
plt.text(1.92, 22.6, '***', fontsize = 12)
bar.hlines(y=22.5, xmin=1.72, xmax=2.25, colors='k')

#plt.annotate("Mean: $19.79", xy=(2.5, 19.79), xytext=(2.7, 25), arrowprops=dict(facecolor='black', shrink=0.05))

#bar.set_yticklabels(bar.get_yticks(), size=10)


# tick length is the length
bar.tick_params(direction='out', length = 3.5, width = 0.9, color='0.2')

plt.savefig('bar_test.png', dpi=300, facecolor=bar.get_facecolor(), orientation = 'portrait')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df["species"] = pd.Categorical.from_codes(iris.target, iris.target_names)


# Create subplots for each feature
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))

for i, ax in enumerate(axes.flat): # axes.flatten()
    sns.boxplot(x=df.species, y=df.iloc[:, i], ax = ax)
    ax.set_title(f'Boxplot for {df.columns[i]}')
    ax.set_xlabel('Species')
    ax.set_ylabel(df.columns[i])
    ax.set_xticks([0, 1, 2])
    ax.spines[['right', 'top']].set_visible(False)


# Adjust layout
plt.tight_layout()
plt.show()


```

```python
from seaborn.widgets import color_palette
plt.figure(figsize=(10,6))
color_palette =sns.color_palette("YlGnBu")
sns.heatmap(df.corr(),annot=True, vmax=1.5, vmin=0,cmap=color_palette, linewidths=0.1)
plt.plot()
```

```python
sns.pairplot(df, hue="species", size=3)
plt.show();
```
